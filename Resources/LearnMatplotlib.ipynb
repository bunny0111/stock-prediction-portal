```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Types of plots

### Line plots

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 40, 50]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True)
```

## Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(months, revenue)
plt.title('Monthly Revenue Bar Chart');
plt.xlabel('Months');
plt.ylabel('Revenue');
plt.grid(True)
```

### Histogram

## Scatter plot

```python
heights = [65, 70, 68, 72, 66]
weights = [158, 160, 155, 180, 145]
```

### Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label='original_price')
plt.plot(days, predicted_price, 'r-o', label='predicted_price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

## Subplot

```python
# First subplot - Origianl Price
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Origianl Price')
plt.title('Origianl prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Second subplot - Predicted Price
plt.subplot(3,1,2)
plt.plot(days, original_price, 'r-o', label='Predicted Price')
plt.title('Predicted prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Third subplot - Difference between Original and Predicted Price
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days, price_difference, color='purple', label='Predicted Price')
plt.title('Difference between Original and Predicted Price')
plt.xlabel('Days')
plt.ylabel('Price Difference')
plt.legend()

plt.tight_layout()
```

### Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='original_price')
plt.plot(days, predicted_price, 'r-o', label='predicted_price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.savefig('example_prediction.png', dpi=300)
```
